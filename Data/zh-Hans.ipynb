```python
import pandas as pd

df = pd.read_csv('yue_en.csv', encoding='utf-8')
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
from translator import translate

yue1 = df['Cantonese'][:1415//2].to_list()
yue2 = df['Cantonese'][1415//2:].to_list()
yue2
#translate(yue[:5], 'yue', ['zh-Hans'])
zh1 = translate(yue1, 'yue', ['zh-Hans'])
zh2 = translate(yue2, 'yue', ['zh-Hans'])
```

```python
zh = []

for z in zh1:
    zh.append(z[0])

for z in zh2:
    zh.append(z[0])

df['Simplified Chinese'] = zh
df.to_csv('yue_en_zh.csv', encoding='utf-8')
df
```
