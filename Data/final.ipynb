```python
import pandas as pd

word = pd.read_csv('word.csv').drop(columns=['Unnamed: 0'])
common = pd.read_csv('Most Common Cantonese Words.csv').drop(columns=['Sentence1', 'Meaning1', 'Jyutping1', 'Sentence2', 'Meaning2', 'Jyutping2', 'Sentence3', 'Meaning3', 'Jyutping3', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16'])
word = word.append(common)
word = pd.concat(g for _, g in word.groupby("Group")).set_index('Cantonese')
word
```

```python
word.to_csv('final_word.csv')
```
