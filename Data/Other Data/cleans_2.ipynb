```python
import pandas as pd

word = pd.read_csv('word.csv').drop(columns=['Unnamed: 0'])
word = pd.concat(g for _, g in word.groupby("Group"))
word[:10]
```

```python
duplicate = pd.concat(g for _, g in word.groupby(["Cantonese", "Part of Speech"]) if len(g) > 1)
duplicate.to_csv('duplicate.csv')
```

```python
duplicate = pd.read_csv('duplicate.csv')
word = word.drop_duplicates(subset=["Cantonese", "Part of Speech"], keep=False).append(duplicate)
word
```

```python
word.to_csv('word.csv')
```

```python
import pycantonese
import pandas as pd

word = pd.read_csv('word.csv').drop(columns=['Sentence1', 'Meaning1', 'Jyutping1', 'Sentence2', 'Meaning2', 'Jyutping2', 'Sentence3', 'Meaning3', 'Jyutping3', 'Unnamed: 0.1', 'Unnamed: 0.1.1', 'Unnamed: 0.1.1.1', 'Unnamed: 0.1.1.1.1', 'Unnamed: 0', 'Unnamed: 0.1.1.1.1.1'])

word
```

```python
import numpy as np
import re

for i in range(word.shape[0]):
    if pd.isnull(word.iloc[i][4]):
        word.iloc[i][4] = pycantonese.characters_to_jyutping(word.iloc[i][0])[0][1]
    else: 
        word.iloc[i][4] = re.sub('\s', '', word.iloc[i][4])

word
```

```python
word.to_csv('word.csv')
```
