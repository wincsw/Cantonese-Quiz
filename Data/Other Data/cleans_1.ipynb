```python
import pandas as pd

group = pd.read_csv("basic_words(group).csv").drop(columns=["Simplified Chinese"])
common = pd.read_csv("Most Common Cantonese Words.csv").drop(columns=["Frequency"])
common.head()
```

```python
common.rename(columns={'Part of speech': 'Group'}, inplace=True)
common.head()
```

```python
word = pd.concat([common,group], axis=0)
word.shape
```

```python
list(word['Group'].unique())
```

```python
import re
verb = word.loc[(word['Group'] == 'Verbs of motion') | (word['Group'] == 'Verbs of perception') | (word['Group'] == 'Verb') | (word['Group'] == 'Verbs')]

english = list(verb['English'].copy())
for e in range(len(english)):
    if re.search("\s/\s", english[e]):
        english[e] = re.sub("\s/\s", ", ", english[e])
    if re.search(";", english[e]):
        english[e] = re.sub(";", ",", english[e])

    if re.search("^to ", english[e]):
        continue
    temp = english[e].split(', ') 
    for t in range(len(temp)):
        temp[t] = "to " + temp[t]
    english[e] = ", ".join(temp)


verb = verb.assign(English=english)
verb = verb.assign(Group=['Verb'] * verb.shape[0])
verb
```

```python
cantonese = verb['Cantonese']
#verb[cantonese.isin(cantonese[cantonese.duplicated()])]
pd.concat(g for _, g in verb.groupby("Cantonese")).to_csv("verb.csv")
```
