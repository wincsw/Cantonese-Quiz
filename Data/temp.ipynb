```python
import pandas as pd

df = pd.read_csv("basic_words(group).csv")
df.head()
```

```python
word = df.drop(columns=['Group'])
word
```

```python
filter = (word['Cantonese'].str.len() == 10)
sorting = word['Cantonese'].str.len().sort_values().index
word = word.reindex(sorting)
word.drop_duplicates()
word = word.reset_index().drop(columns=['index'])
canto = word['Cantonese']
word.drop(labels=['Cantonese'], axis=1,inplace = True)
word.insert(0, 'Cantonese', canto)
word
```

```python
word[['Cantonese', 'English']].to_csv('basic_words(en).csv')
```

```python
zh = word[['Cantonese', 'Simplified Chinese']]
filter = (word['Cantonese'] != word['Simplified Chinese'])
zh = zh.loc[filter].reset_index().drop(columns=['index'])
zh
```

```python
zh.to_csv('basic_words(zh).csv')
```
